```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('submit2.txt', 'r') as f:
    edge = [line.strip() for line in f]
for i in range(len(edge)):
    edge[i] = [int(edge[i].split()[0]), int(edge[i].split()[1])]
edge
```

```python
G = nx.Graph()
G.add_edges_from(edge)
```

```python
nx.draw(G)
```
